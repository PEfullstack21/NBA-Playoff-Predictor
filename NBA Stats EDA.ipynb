# Exploring Data with Pandas

```python
import pandas as pd

df = pd.read_csv(r'C:\Users\anooj\Desktop\ga-data\NBA-Playoff-Predictor\NBA Total Stats.csv')
df = pd.DataFrame(df)
df.head()
```

```python
#print the last 3 rows
df.tail(3)
```

```python
# examine the datatypes of the columns
# note that these were automatically inferred by pandas upon import!
pd.DataFrame(df.dtypes, columns=['Data Type'])
```

```python
# displaying the index as it sits (auto-generated upon import)
df.index
```

```python
# also note that our auto-generated index has no name
print(df.index.name)
```

```python
# Here we are looking at three columns;
# the one on the left is the index (automatically generated upon import by pandas)
# 'Rk' is our PK (primary key) from our imported table. 'Name' is a data column.
# Notice that the generated index starts at zero and our PK starts at 1.
pd.DataFrame(df.head(3)[['Rk', 'Team']])
```

```python
#new dataframe to slice the data set for one years worth of data
year_df = df[df['Season'] == '20-21']
year_df
```

```python
# let's sort by steals, descending
year_df.sort_values(by='STL', ascending=False).head(3)
```
